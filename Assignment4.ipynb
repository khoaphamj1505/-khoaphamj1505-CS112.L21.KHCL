# Sửa Hàng Rào
- Link: https://khmt.uit.edu.vn/wecode/cs112.2021/assignment/6/6

- Abstraction: 
    * Cho hai mảng A gồm n phần tử 
$a_1, a_2 ... a_n$ và mảng B gồm m phần tử 
$b_1, b_2 ... b_m$.
    * Gọi $F(A)$ là giá trị nhỏ nhất của mảng A sau khi mảng A bị biến đổi
    * Chọn 2 tập gồm k phần tử 
$S_1 = {i_1, i_2 ... i_k}| (1 \leq i_1 < i_2 < ... < i_k \leq n)$ và 
$S_2 = {j_1, j_2 ... j_k}| (1 \leq j_1 < j_2 < ... < j_k \leq m)$ sau đó lần lượt thay thế các giá trị 
$a_{i_x} = a_{i_x} + b_{j_x} (i_x \in S1, j_x \in S2)$ để cho 
$F(A)$ đạt giá trị lớn nhất.
    * Output:

        $max(F(A))$ $k$

        $i_1$  $j_1$

        $i_2$  $j_2$

        ...

        $i_k$ $j_k$
    * Điều kiện:

        $0 \leq k \leq n$.

        $1 \leq n, m \leq 10^5 $

        $1 \leq a_i, b_j \leq 10^8 | (i \in [1; n], j \in [1; m])$.

    

* Pattern Recognition:
    * Nếu không tồn tại 2 tập 
$S1 và S2$ để sau khi biến đổi thì 
$min(a_1, a_2 ... a_n) \geq x$ thì ta chắc chắn rằng 
$1 \leq max(F(A)) \leq x$.
    * Ngược lại, Nếu tồn tại tập 
$S1 và S2$ để sau khi biến đổi thì 
$min(a_1, a_2 ... a_n) \geq x$ thì ta chắc chắn rằng 
$x \leq max(F(A)) \leq 2 * 10^8$. (Do giá trị của các phần tử trong 2 mảng A và B không vuợt quá $10^8$)

    * Để kiểm tra liệu có tồn tại 2 tập 
$S1, S2$ để sau khi biến đổi thì $F(A) \geq x$ ta có thể  thực hiện trong thời gian tuyến tính 

    * Có thể thực hiện tìm kiếm nhị phân để tìm $max(F(A))$

* Algorithms Design:

    - B1: Thiết kế hàm check(height) để kiểm tra có tồn tại 2 tập $S1, S2$ để $F(A) \geq height$ hay không
        * check(height) trả về True nếu tồn tại $S1, S2$
        * Ngược lại trả về 0
    - B2: Khởi tạo giá trị 

        * l = 1

        * r = $2* 10^8$ 

        * mid = (l + r) // 2
    - B3: Nếu l < r:
        - B3.1: Nếu check(mid) == True thì l = mid + 1, ngược lại thì r = mid
        - B3.2: Cập nhật  mid = (l + r) // 2

    - B4: Nếu l < r thì lặp lại B3

    - B5: In ra màn hình giá trị l - 1
    - B6: Tìm và Liệt kê kích thước cũng như từng phần tử của S1 và S2 để F(A) đạt giá trị bằng l - 1

- Time Complexity của thuật toán: $O(NLog(N))$
      


   

````python
def check(mid, n, m, a, b):
	low_val = [x for x in a if x < mid]
	L = len(low_val)
	if L == 0:
		return 1
	i = 0

	for val in b:
		if val + low_val[i] >= mid:
			i += 1

		if  i == L:
			return 1

	return 0

def get_answer(max_height, n, m, a, b):
	low_id = [i for i in range(n) if a[i] < max_height]

	L = len(low_id)
	print(max_height, L)
	if L == 0:
		return
	i = 0

	ans = ''
	for j in range(m):
		
		if b[j] + a[low_id[i]] >= max_height:
			# print(low_id[i] + 1, j + 1)
			ans += str(low_id[i] + 1) +  ' ' + str(j + 1) + '\n'
			i += 1

		if L == i:
			print(ans, end='')
			return 

def main():
	n = int(input())
	a = [int(x) for x in input().split()]

	m = int(input())
	b = [int(x) for x in input().split()]

	l = 1
	r = 2* (10**9)

	while l < r :
		mid = (l + r) // 2
		if check(mid, n, m, a, b) :
			l = mid + 1
		else: 
			r = mid 

	max_height = l - 1
	get_answer(max_height, n, m, a, b)


main()
````

# Phân số
- Link: https://khmt.uit.edu.vn/wecode/cs112.2021/assignment/6/13

- Abstraction: 
    * Cho 2 phân số tối giản
$\frac{a}{b}$ và 
$\frac{c}{d}$
$(\frac{a}{b} \neq \frac{c}{d})$ . Một thao tác biến đổi sẽ tăng tử số và mẫu số  của phân số 
$\frac{a}{b}$ lên một đơn vị sau đó rút gọn phân số đó.

    * Hỏi sau ít nhất bao nhiêu thao tác thì 
$\frac{a}{b} = \frac{c}{d}$

- Pattern Recognition:
    * Thao tác tăng tử và mẫu số của một phân số lên 1 đơn vị sẽ làm tăng giá trị của phân số ban đầu 
    * Theo nhận xét phía trên nếu
$\frac{a}{b} > \frac{c}{d}$ thì không thể biến đổi để 
$(\frac{a}{b} = \frac{c}{d})$

        => Vì vậy lời giải cho bài toán này là liên tục thực hiện thao tác cho đến khi 
$(\frac{a'}{b'} >= \frac{c}{d})$

    * Vì giá trị của 
$a, b, c, d$ đều nằm trong khoảng 
$10^5$ nên ta chỉ cần lặp tuyến là có thể giải quyết bài toán.
- Algorithm Design:
    * Nhập 4 số a, b, c, d
    * Nếu 
$ a * d > b * c$ thì in 0 ra màn hình và kết thúc chương trình
    * Ngược lại, thực hiện vòng lặp sau:
``` python
    # Tạo 1 biến đếm số thao tác được thực thi
    cnt = 0

     while a * d < b * c:
        # Tăng giá trị của a và b lên 1 đơn vị
            a += 1
            b += 1

        # Tính ước chung lớn nhất của a và b
            s = gcd(a, b)

        # chia a và b cho ước chung lớn nhất để rút gọn phân số a/b
            a //= s
            b //= s
```

    - Vòng lặp kết thúc khi 
$\frac{a}{b} >= \frac{c}{d}$. Sau khi vòng lặp kết thúc, kiểm tra xem hai phân số có bằng nhau hay không. 
        * Nếu có thì in ra giá trị biến cnt lưu số thao tác được thực thi
        * Nếu không thì in ra giá trị 0



````python
from math import gcd

if __name__ == '__main__':
	a = int(input())
	b = int(input())
	c = int(input())
	d = int(input())
	cnt = 0

	if a * d > b * c:
		print(0)
	else:
		while a * d < b * c:	
			cnt += 1
			a = a + 1
			b = b + 1

			_gcd = gcd(a, b)
			a //= _gcd
			b //= _gcd

		if a * d != b * c:
			print(0)
		else:
			print(cnt)
````
